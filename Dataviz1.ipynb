```
import matplotlib.pyplot as plt
import numpy
```

```
x1 = numpy.arange(0,10,1)
x2 = numpy.linspace(0, 5, 10)
plt.plot(x1, x1, label="linear")
plt.plot(x1, x2**2, label="quadratic")

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```

```
from scipy.interpolate import interp1d
fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x1, x2)
ax1.set_title('LINEAR PLOT')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')


x = numpy.arange(0, 4*numpy.pi, 0.1)
y = numpy.cos(x)
z = numpy.sin(x)
ax2.plot(x, y, x, z)
ax2.set_title('COSINE & SINE PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

```
x = numpy.random.randn(10)
y = numpy.random.randn(10)
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```
