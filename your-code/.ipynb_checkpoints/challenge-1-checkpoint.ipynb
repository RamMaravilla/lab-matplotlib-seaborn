```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
#fig, (ax1, ax2) = plt.subplots(2)
#fig.suptitle('Vertically stacked subplots')
#ax1.plot(x, y)
#ax2.plot(x, -y)


fig, axes = plt.subplots(2)
fig.suptitle("Prueba")
axes[0].plot(x, y)   #Y es el eje del primero
axes[1].plot(x, z)   #Z del segund
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here

#Hacer dos líneas y después en la misma gráfica

#1 Definir variables 
x = np.arange(0,100)
y = x^2
z = x**2  # ** significa exponencial 

#2 Decirle que las plotee
plt.plot(x, y, label = "Cuadratic")
plt.plot(x, z, label = "Exponential")

#3 Le pones nombre a los ejes
plt.xlabel('Eje x')
plt.ylabel('Eje y')

#4 Le pones titutlo, legenga y que lo imprima
plt.title("Cuadratic vs Exponential")
#plt.legend()
#plt.show()

```
